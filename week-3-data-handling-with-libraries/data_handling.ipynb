```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load CSV file
df = pd.read_csv("data/sample.csv")

print("=== Full Dataset ===")
print(df)

# Basic Info
print("\n=== Dataset Info ===")
print(df.info())

# Descriptive Statistics
print("\n=== Descriptive Statistics ===")
print(df.describe())

# Filtering: only CSE Department
cse_students = df[df["Department"] == "CSE"]
print("\n=== CSE Students ===")
print(cse_students)

# Sorting by CGPA
sorted_df = df.sort_values(by="CGPA", ascending=False)
print("\n=== Sorted by CGPA ===")
print(sorted_df)

# --- Visualizations ---

# Bar Chart: CGPA per Student
plt.figure(figsize=(6,4))
plt.bar(df["Name"], df["CGPA"])
plt.title("CGPA of Students")
plt.xlabel("Student Name")
plt.ylabel("CGPA")
plt.tight_layout()
plt.savefig("cgpa_bar_chart.png")
plt.show()

# Scatter Plot: Age vs CGPA
plt.figure(figsize=(6,4))
plt.scatter(df["Age"], df["CGPA"])
plt.title("Age vs CGPA")
plt.xlabel("Age")
plt.ylabel("CGPA")
plt.tight_layout()
plt.savefig("age_cgpa_scatter.png")
plt.show()

# Heatmap of Correlation
plt.figure(figsize=(5,4))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.savefig("correlation_heatmap.png")
plt.show()
```
